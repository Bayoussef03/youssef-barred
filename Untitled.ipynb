```python
import pandas as pd

file_path = 'combined_tables.xlsx'
base_y = pd.read_excel(file_path)

print("Avant la suppression des colonnes :")
print(base_y.head())

cols_to_drop = ['Position', 'Pos']
base_y.drop(cols_to_drop, axis=1, inplace=True)

nouveau_chemin = 'base_y_sans_positions.xlsx'
base_y.to_excel(nouveau_chemin, index=False)

print("\nAprès la suppression des colonnes :")
print(base_y.head())
```

```python
base_y = base_y[(base_y['Montant de transfert'] != '-') & (base_y['Montant de transfert'] != '?')]

base_y.reset_index(drop=True, inplace=True)

nouveau_chemin = 'base_y_sans_positions_et_montant.xlsx'
base_y.to_excel(nouveau_chemin, index=False)

base_y.head()
```

```python
base_y = base_y.dropna(subset=['joueurs'])

base_y.reset_index(drop=True, inplace=True)

nouveau_chemin = 'base_y_sans_positions_montant_joueurs_nuls.xlsx'
base_y.to_excel(nouveau_chemin, index=False)

nombre_lignes_restantes_apres_suppression = base_y.shape[0]
print(f"Il reste {nombre_lignes_restantes_apres_suppression} lignes après la suppression des joueurs nuls.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fichier_donnees = 'base_y_montant_propre_chiffres_virgule.xlsx'
donnees = pd.read_excel(fichier_donnees)

donnees['Montant de transfert'] = donnees['Montant de transfert'].str.replace(',', '')

donnees['Montant de transfert'] = pd.to_numeric(donnees['Montant de transfert'])

donnees_triees = donnees.sort_values('Montant de transfert', ascending=False)

top_10_chers = donnees_triees.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_chers['joueurs'], top_10_chers['Montant de transfert'], color='skyblue')
plt.xlabel('Montant de transfert (en millions)')
plt.ylabel('Joueurs')
plt.title('Top 10 des transferts les plus chers par joueur')
plt.gca().invert_yaxis()  # Inverser l'axe y pour avoir le joueur le plus cher en haut
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
top_10_equipes = donnees.groupby('equipe')['Montant de transfert'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10, 6))
plt.barh(top_10_equipes['equipe'], top_10_equipes['Montant de transfert'], color='lightgreen')
plt.xlabel('Montant total de transfert (en millions)')
plt.ylabel('Équipes')
plt.title('Top 10 des montants de transfert par équipe')
plt.gca().invert_yaxis()  # Inverser l'axe y pour avoir l'équipe avec le montant le plus élevé en haut
plt.tight_layout()

plt.show()
```

```python
top_10_equipes = donnees.groupby('equipe')['Montant de transfert'].sum().nlargest(10).reset_index()

top_10_equipes['Montant de transfert'] = (top_10_equipes['Montant de transfert'] / 100).astype(int)

plt.figure(figsize=(10, 6))
plt.barh(top_10_equipes['equipe'], top_10_equipes['Montant de transfert'], color='lightgreen')
plt.xlabel('Montant total de transfert (en millions)')
plt.ylabel('Équipes')
plt.title('Top 10 des montants de transfert par équipe (en millions)')
plt.gca().invert_yaxis()  # Inverser l'axe y pour avoir l'équipe avec le montant le plus élevé en haut
plt.tight_layout()

plt.show()
```

```python
top_10_transferts_par_equipe = donnees.groupby('equipe').apply(lambda x: x.nlargest(10, 'Montant de transfert')).reset_index(drop=True)

plt.figure(figsize=(10, 6))
for equipe, data in top_10_transferts_par_equipe.groupby('equipe'):
    plt.barh(data['joueurs'], data['Montant de transfert'], label=equipe)

plt.xlabel('Montant de transfert')
plt.ylabel('Joueurs')
plt.title('Top 10 des transferts les plus chers par équipe')
plt.legend(title='Équipe', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```

```python
top_10_transferts_par_equipe = donnees.groupby('equipe').apply(lambda x: x.nlargest(10, 'Montant de transfert')).reset_index(drop=True)

plt.figure(figsize=(10, 6))
for equipe, data in top_10_transferts_par_equipe.groupby('equipe'):
    plt.barh([equipe] * len(data), data['Montant de transfert'], label=equipe)

plt.xlabel('Montant de transfert')
plt.ylabel('Équipes')
plt.title('Top 10 des transferts les plus chers par équipe')
plt.legend(title='Équipe', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```

```python
top_10_transferts_par_equipe = donnees.groupby('equipe').apply(lambda x: x.nlargest(10, 'Montant de transfert')).reset_index(drop=True)

top_10_transferts_par_equipe['Montant de transfert'] = (top_10_transferts_par_equipe['Montant de transfert'] // 100)

plt.figure(figsize=(10, 6))
for equipe, data in top_10_transferts_par_equipe.groupby('equipe'):
    plt.barh([equipe] * len(data), data['Montant de transfert'], label=equipe)

plt.xlabel('Montant de transfert (en millions)')
plt.ylabel('Équipes')
plt.title('Top 10 des transferts les plus chers par équipe')
plt.legend(title='Équipe', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```

```python
total_transferts_par_equipe = donnees.groupby('equipe')['Montant de transfert'].sum().reset_index()

total_transferts_par_equipe = total_transferts_par_equipe.sort_values('Montant de transfert', ascending=False)

top_10_transferts_par_equipe = donnees[donnees['equipe'].isin(total_transferts_par_equipe['equipe'].head(10))]

top_10_transferts_par_equipe['Montant de transfert'] = (top_10_transferts_par_equipe['Montant de transfert'] // 100)

plt.figure(figsize=(10, 6))
for equipe, data in top_10_transferts_par_equipe.groupby('equipe'):
    plt.barh([equipe] * len(data), data['Montant de transfert'], label=equipe)

plt.xlabel('Montant de transfert (en millions)')
plt.ylabel('Équipes')
plt.title('Top 10 des transferts les plus chers par équipe')
plt.legend(title='Équipe', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```
